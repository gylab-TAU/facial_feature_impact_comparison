```python
import pandas as pd
import torch
import pickle as pkl
import glob
```

```python
root = r"/home/administrator/experiments/familiarity/finetuned_B/vgg16/results/activations"
```

```python
pairs = pd.read_csv(r"/home/administrator/datasets/high_low_ps_images/image_pairs_lists/diff_pairs.txt", sep=" ", header=None, index_col=False)
pairs
```

```python
import numpy as np
distances = pd.DataFrame(columns=['pair', 'layer', 'dist'])
dist_calc = np.linalg.norm #torch.nn.PairwiseDistance()
layers = ['conv1_MaxPool2d(kernel_size=2, stride=2, padding=0, dilation=1, ceil_mode=False)',
         'conv2_MaxPool2d(kernel_size=2, stride=2, padding=0, dilation=1, ceil_mode=False)',
         'conv3_MaxPool2d(kernel_size=2, stride=2, padding=0, dilation=1, ceil_mode=False)',
         'conv4_MaxPool2d(kernel_size=2, stride=2, padding=0, dilation=1, ceil_mode=False)',
         'conv5_MaxPool2d(kernel_size=2, stride=2, padding=0, dilation=1, ceil_mode=False)',
         'fc6_ReLU(inplace=True)',
         'fc7_ReLU(inplace=True)',
         'fc8_Linear(in_features=4096, out_features=8849, bias=True)']
for i in range(len(pairs)):
    for l in layers:
        im_name1 = pairs.iloc[i][0]
        im_name2 = pairs.iloc[i][1]
        path1 = f'{root}/{im_name1}/{l}.pkl'
        path2 = f'{root}/{im_name2}/{l}.pkl'
        with open(path1, 'rb') as pkld_img1:
            with open(path2, 'rb') as pkld_img2:
                im1 = pkl.load(pkld_img1)
                im2 = pkl.load(pkld_img2)
                distances = distances.append(pd.DataFrame({'pair': [f'{im_name1}, {im_name2}'], 'layer': [l.split('_')[0]], 'dist': [dist_calc(im1['rep'].numpy() - im2['rep'].numpy())]}))
```
